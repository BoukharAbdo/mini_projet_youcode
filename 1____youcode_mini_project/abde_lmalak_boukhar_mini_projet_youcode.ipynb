<h5>Calculer la médiane et la moyenne des colonnes :<br>
○ « Âge »<br>
○ « Dépenses des clients »
</h5>

<h6>Convertis le type de 'âge' en entier (int) et 'Dépenses des clients' en flottant (float) avec le calcul</h6>

```python
# Nettoyer les espaces en trop dans les noms des colonnes
data.columns = data.columns.str.strip()
```

```python
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Age'] = data['Age'].astype('Int64')
data['Dépenses des clients'] = pd.to_numeric(data['Dépenses des clients'], errors='coerce')
```

```python
# Calculer la médiane et la moyenne de l'âge
age_median = data['Age'].median()
age_mean = data['Age'].mean()

print(age_median)
print(age_mean)
```

```python
# Calculer la médiane et la moyenne des dépenses des clients
expenses_median = data['Dépenses des clients'].median()
expenses_mean = data['Dépenses des clients'].mean()

print(expenses_median)
print(expenses_mean)
```

```python
# Afficher les statistiques descriptives après conversion
print("\nStatistiques descriptives après conversion des âges négatifs en positifs :")
print(data['Age'].describe())
```

```python
# Afficher les statistiques descriptives après conversion
print("\nStatistiques descriptives après conversion des dépenses des clients négatifs en positifs :")
print(data['Dépenses des clients'].describe())
```

```python
print(data.info())
```

```python
# Remplacement par la médiane pour "Âge" et "Dépenses des clients"
age_median = data['Age'].median()
depenses_median = data['Dépenses des clients'].median()

data['Age'] = data['Age'].fillna(age_median)
data['Dépenses des clients'] = data['Dépenses des clients'].fillna(depenses_median)
print(data.info())
```

```python
# Vérifier les valeurs nulles restantes
print("Vérification des valeurs nulles après remplacement :")
print(data.isnull().sum())
```

```python
# Trouver la valeur la plus fréquente (mode) dans la colonne "Genre"
mode_genre = data['Genre'].mode()[0]

# Remplacer les valeurs nulles dans la colonne "Genre" par la valeur la plus fréquente
data['Genre'] = data['Genre'].fillna(mode_genre)

# Vérifier les valeurs nulles restantes
print("Vérification des valeurs nulles après remplacement :")
print(data.isnull().sum())
```

```python
print(data.info())
```

<h5>Créer une visualisation des données du graphique à barres qui montre les dépenses
des clients par pays</h5>

```python
import matplotlib.pyplot as plt

depenses_par_pays = data.groupby('Pays')['Dépenses des clients'].sum()

plt.figure(figsize=(10, 6))
depenses_par_pays.plot(kind='bar', color='skyblue')

plt.title('Dépenses des clients par pays')
plt.xlabel('Pays')
plt.ylabel('Dépenses totales (€)')

plt.show()
```

<h6>Nettoyer les lignes avec moins de 10 € de dépenses client : supprimez toutes les
